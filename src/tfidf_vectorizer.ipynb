```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse
import pickle
```

```python
# load data
data = pd.read_csv('Datasets/recipes_food_com_revised.csv')
data.head()
```

```python
# initialize TfidfVectorizer and fit on ingredients
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(data['NLP_Ingredients'])
```

```python
# save the TF-IDF matrix and the vectorizer
scipy.sparse.save_npz('tfidf_matrix.npz', tfidf_matrix)  # Save the sparse matrix
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```
